## Download data

```python
def retrieve_data(year, month, days, client):
    dataset = "reanalysis-era5-single-levels"
    request = {
        "product_type": ["reanalysis"],
        "variable": ["2m_temperature"],
        "year": [year],
        "month": [month],
        "day": days,
        "time": [
            "00:00", "01:00", "02:00",
            "03:00", "04:00", "05:00",
            "06:00", "07:00", "08:00",
            "09:00", "10:00", "11:00",
            "12:00", "13:00", "14:00",
            "15:00", "16:00", "17:00",
            "18:00", "19:00", "20:00",
            "21:00", "22:00", "23:00"
                ],
        "data_format": "netcdf",
        "download_format": "unarchived",
        "area": [71.5, 27, 41, 60]
        }

    directory_path = f'data/year_{year}/month_{month}'
    os.makedirs(directory_path, exist_ok=True)
    
    target = 'data/year_{year}/month_{month}/ERA5_2m_{year}_{month}.nc'.format(year=year, month=month)
    client.retrieve(dataset, request, target)

def download_data(init_year, client):
    init_month = 1

    for i in range(0, 11):
        cur_year = str(init_year + i)
        for j in range(0, 12):
            if init_month + j < 10:
                cur_month = "0" + str(init_month + j)
            else:
                cur_month = str(init_month + j)
            numberOfDays = calendar.monthrange(init_year + i, init_month + j)[1]
            # create list of days in the given month
            cur_days = []
            for k in range(1, numberOfDays+1):
                if k < 10:
                    cur_days.append("0" + str(k))
                else:
                    cur_days.append(str(k))
            retrieve_data(cur_year, cur_month, cur_days, client)
    
```

## Data preprocessing

Initial datasets. Average daily temperature difference for the first day of 2010 is calculated separately (there is a faster way, but i did not have time to redo it)

```python
# Set the initial month to January for data processing
init_month = 1

# Load datasets for December 2009 and January 2010, initializing final_dataset and prev_dataset for further processing
december_2009 = xr.open_dataset("data/year_2009/month_12/ERA5_2m_2009_12.nc", engine="netcdf4")
final_dataset = xr.open_dataset("data/year_2010/month_01/ERA5_2m_2010_01.nc", engine="netcdf4")
prev_dataset = xr.open_dataset("data/year_2010/month_01/ERA5_2m_2010_01.nc", engine="netcdf4")

# Convert the temperature variable 't2m' from Kelvin to Celsius, storing it as 'avg_daily_temp' in each dataset
december_2009['avg_daily_temp'] = december_2009['t2m'] - 273.15
final_dataset['avg_daily_temp'] = final_dataset['t2m'] - 273.15
prev_dataset['avg_daily_temp'] = prev_dataset['t2m'] - 273.15

# Remove the original 't2m' variable since temperature is now represented in Celsius
december_2009 = december_2009.drop_vars('t2m')
final_dataset = final_dataset.drop_vars('t2m')
prev_dataset = prev_dataset.drop_vars('t2m')

# Aggregate data by year, month, and day, and compute daily means for each dataset
december_2009 = december_2009.groupby(["valid_time.year", "valid_time.month", "valid_time.day"]).mean()
final_dataset = final_dataset.groupby(["valid_time.year", "valid_time.month", "valid_time.day"]).mean()
prev_dataset = prev_dataset.groupby(["valid_time.year", "valid_time.month", "valid_time.day"]).mean()

# Calculate the daily temperature difference and store it as 'avg_daily_temp_diff' in each dataset
final_dataset['avg_daily_temp_diff'] = final_dataset['avg_daily_temp'].diff(dim="day")
prev_dataset['avg_daily_temp_diff'] = prev_dataset['avg_daily_temp'].diff(dim="day")

# Set December 2009's coordinates to 2010/January to align with January 2010 data for calculations
december_2009.coords['year'] = xr.DataArray([2010] * len(december_2009.year), dims="year")
december_2009.coords['month'] = xr.DataArray([1] * len(december_2009.month), dims="month")
december_2009 = december_2009["avg_daily_temp"].loc[
            dict(latitude=slice(None), longitude=slice(None), year=slice(None), month=slice(None), day=31)
        ]

# Extract the temperature on the first day of January 2010 for temperature difference calculations
january_2010 = prev_dataset["avg_daily_temp"].loc[
            dict(latitude=slice(None), longitude=slice(None), year=slice(None), month=slice(None), day=1)
        ]

# Adjust the daily temperature difference for January 1 based on the temperature change from December 31
prev_dataset["avg_daily_temp_diff"].loc[
            dict(latitude=slice(None), longitude=slice(None), year=slice(None), month=slice(None), day=1)
] = january_2010 - december_2009

final_dataset["avg_daily_temp_diff"].loc[
            dict(latitude=slice(None), longitude=slice(None), year=slice(None), month=slice(None), day=1)
] = january_2010 - december_2009

# Define a 'sudden_temp_change' variable to flag temperature changes of 5C or more
final_dataset['sudden_temp_change'] = xr.where(
    abs(final_dataset['avg_daily_temp_diff']) >= 5, 1, 0
)
prev_dataset['sudden_temp_change'] = xr.where(
    abs(prev_dataset['avg_daily_temp_diff']) >= 5, 1, 0
)

# Initialize the number of days in January as 31
init_numberOfDays = 31
```

Loop over the years and months to correctly calculate the average daily temperature for the next day of next month and put it to the final_dataset

```python
# Loop through each year from 2010 to 2020 (i = 0 to 10)
for i in range(0, 11):
    cur_year = str(init_year + i)  # Set the current year as a string

    # Loop through each month from January to December (j = 0 to 11)
    for j in range(0, 12):
        # Skip the first iteration as January 2010 data is already loaded
        if j == 0 and i == 0:
            continue

        # Format the current month as a two-digit string (e.g., '01' for January)
        if init_month + j < 10:
            cur_month = "0" + str(init_month + j)
        else:
            cur_month = str(init_month + j)

        # Determine the number of days in the current month
        cur_numberOfDays = calendar.monthrange(init_year + i, init_month + j)[1]

        # Load the dataset for the current month and convert temperatures from Kelvin to Celsius
        cur_dataset = xr.open_dataset(f"data/year_{cur_year}/month_{cur_month}/ERA5_2m_{cur_year}_{cur_month}.nc", engine="netcdf4")
        cur_dataset['avg_daily_temp'] = cur_dataset['t2m'] - 273.15

        # Drop the original 't2m' variable, group data by year, month, day, and calculate daily mean temperatures
        cur_dataset = cur_dataset.drop_vars('t2m')
        cur_dataset = cur_dataset.groupby(["valid_time.year", "valid_time.month", "valid_time.day"]).mean()

        # Compute daily temperature differences for the current month
        cur_dataset['avg_daily_temp_diff'] = cur_dataset['avg_daily_temp'].diff(dim="day")

        # Update coordinates in prev_dataset to align with the current year and month for continuity
        prev_dataset.coords['year'] = xr.DataArray([int(cur_year)] * len(prev_dataset.year), dims="year")
        prev_dataset.coords['month'] = xr.DataArray([j + 1] * len(prev_dataset.month), dims="month")

        # Get the temperature on the last day of the previous month
        last_prev_month = prev_dataset["avg_daily_temp"].loc[
            dict(latitude=slice(None), longitude=slice(None), year=slice(None), month=slice(None), day=init_numberOfDays)
        ]

        # Get the temperature on the first day of the current month
        first_cur_month = cur_dataset["avg_daily_temp"].loc[
            dict(latitude=slice(None), longitude=slice(None), year=slice(None), month=slice(None), day=1)
        ]

        # Calculate the temperature difference between the first day of the current month and the last day of the previous month
        cur_dataset["avg_daily_temp_diff"].loc[
            dict(latitude=slice(None), longitude=slice(None), year=slice(None), month=slice(None), day=1)
        ] = first_cur_month - last_prev_month

        # Create a variable 'sudden_temp_change' to identify days with a temperature change of 5C or more
        cur_dataset['sudden_temp_change'] = xr.where(
            abs(cur_dataset['avg_daily_temp_diff']) >= 5, 1, 0
        )

        # Add the current month's dataset to final_dataset, accumulating data month by month
        final_dataset = xr.merge([final_dataset, cur_dataset])

        # Update prev_dataset to the current dataset for use in the next iteration
        prev_dataset = cur_dataset

        # Set the number of days in the current month to use in the next loop iteration
        init_numberOfDays = cur_numberOfDays
```
